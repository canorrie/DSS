```
d1 = open("Day1.lua", "r")
s1 = str(d1.read()).splitlines()
d2 = open("Day2.lua", "r")
s2 = str(d2.read()).splitlines()
d3 = open("Day3.lua", "r")
s3 = str(d3.read()).splitlines()
d4 = open("Day4.lua", "r")
s4 = str(d4.read()).splitlines()
d5 = open("Day5.lua", "r")
s5 = str(d5.read()).splitlines()
d6 = open("Day6.lua", "r")
s6 = str(d6.read()).splitlines()
d7 = open("Day7.lua", "r")
s7 = str(d7.read()).splitlines()
s = [s1,s2,s3,s4,s5,s6,s7]
```

```
'''
INDICES:
Item name: 8 (needs characters removed)
Item ID: 22
Bid: 5
Buyout: 16
Scantime: 7
Item seller: 19 (needs characters removed)
Item quantity: 10
'''

big_list = []
list_number = 1
for x in s:
  the_line = 0
  index = 0
  for i in x:
    if "ropes" in i:
      the_line = index + 1
    index += 1
  the_string = x[the_line].split('{')[2:]
  for i in the_string:
    p = i.split(',')
    q = []
    q.append(p[8])
    q.append(p[22])
    q.append(p[5])
    q.append(p[16])
    q.append(p[7])
    q.append(p[19])
    q.append(p[10])
    q.append(list_number)
    big_list.append(q)   
  list_number += 1        
```

```
df = pd.DataFrame.from_records(big_list, columns=['Item Name', 'Item ID', 'Bid', 'Buyout', 'Scantime', 'Seller', 'Quantity', 'Day'])
```
